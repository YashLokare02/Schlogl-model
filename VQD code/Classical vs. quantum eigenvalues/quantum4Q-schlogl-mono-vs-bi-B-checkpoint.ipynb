## Introduction
This notebook demonstrates how to solve the Schlogl model using quantum computing. We focus on a variational based method known as the variational quantum deflation (VQD) algorithm to obtain the first and second smallest magnitude eigenvalues of the Schlogl operator.

## Variational Quantum Deflation (VQD)
Finding eigevalues of a matrix using the variational quantum deflation (VQD) algorithm. Note that VQD is an extension of the well-known variational quantum eigensolver (VQE). The only distiction is that VQD is capable of computing excited states.

### Hermitian Matrix to Pauli Operator
To diagonalize a Hermitian matrix on a quantum computer, we need to convert it into a Pauli operator, i.e., represent it as a superposition of the tensor products of the Pauli operators, $X$, $Y$, $Z$, and $I$. This is achieved using the formula:
$$
H = a_0 I  + a_1 X + a_2 Y + a_3 Z,
$$
where the coefficients $a_0$ to $a_3$ are given by
$$
a_0 = \frac{\Tr(H \otimes I)}{2^n}\\
a_1 = \frac{\Tr(H \otimes X)}{2^n}\\
a_2 = \frac{\Tr(H \otimes Y)}{2^n}\\
a_3 = \frac{\Tr(H \otimes Z)}{2^n}.\\
$$
Note that $n$ is the number of qubits.

In qiskit, this can be achieved using the in-built function `SparsePauliOp.from_operator(matrix)` from the `qiskit.quantum_info` module. 

After transforming the Hermitian matrix into a sum of Pauli operators, we can run the quantum simulation and variationally optimize the eigenvalues using the classical-quantum hybrid scheme in qiskit's VQD module. The parameters of the quantum circuit are optimized classcially while the eigenvalues are estimated using the VQE quantum algorithm.

### Defining Problem
We will start by redefining our Schlogl operator for eigenvalue evaluation. We will use a smaller matrix for our quantum simulation since it's not feasible to quantum mechanically simulate a large matrix on classical hardware. We will classically evaluate the eigenvalues of the Schlogl operator to use as reference for the quantum results. Since the VQE and VQD algorithms only compute the eigenvalues, we will ignore the eigenvectors from here on. Furthermore, previous numerical tests have shown that there is a correlation between the truncation value, $N$, and the accuracy threshold at different $V$. We found that larger operators are required to acurately determine the eigenvalues when $V$ is large. Since we are focusing on smaller operators as mentioned above, we will use relatively smaller volumes than in our previous experiments (see notebook _classical-schlogl.ipynb_). We will do one extra transformation to our Hermitian matrix to ensure that all eigenvalues are positive. This is important for the VQE to find the smallest magnitude eigenvalues. If we do not perform such a transformation, the VQE returns the lowest (including negative values) eigenvalues. 

Although the Hermitian matrix we used previously, 
\begin{equation}
    \label{eq:schl4}
    Q_H = \begin{pmatrix}
\mathbf{0} &Q  \\
Q^{\dagger} &\mathbf{0} 
\end{pmatrix},
\end{equation}
conserves both the eigenvectors and eigenvalues in a sense, it is not positive definite. Thus, this transformation is not suitable for the VQE algorithm. To construct an Hermitian semi-positive definite operator, we will use the transformation
\begin{equation}
Q_{\mathrm{spd}} = Q\cdot Q^T,
\end{equation}
whose eigenvalues are squares of the original matrix's eigenvalues, such that
$$
\lambda^{Q}_{i} = \sqrt{\lambda^{Q_{\mathrm{spd}}}_{i}}
$$
where $\lambda^{Q}_{i}$ are the eigenvalues of the original Schlogl operator and $\lambda^{Q_{\mathrm{spd}}}_{i}$ are the eigenvalues of the transformed semi-definite matrix.

### Monostable Model Parameters

```python
# monostable Schlogl
a = 0.5
b = 29.5
k1 = 3
k2 = 0.6
k3 = 0.25
k4 = 2.95

# number of qubits
operator_num_qubits = 4
print("Initializing...\n\n","-->",str(operator_num_qubits)+"-qubit matrix")

# Matrix size should be 2^num_qubits
N = 2**operator_num_qubits
print(" --> Matrix size = ",N,"\n...")

# volume array
Vn = np.arange(0.6, 11.0, 0.5)
```

## Exact Simulation

```python
# high convergence threshold
N_exact = 2**5
print("Exact matrix size = ",N_exact)

# volume array
Vn_exact = Vn

# eigenvalue arrays
exact_eigenvalue0 = np.zeros(len(Vn_exact))
exact_eigenvalue1 = np.zeros(len(Vn_exact))

#******************************************
for i, V in enumerate(Vn_exact):
    # birth and death rates
    lambda_fn = lambda n: ((a*k1*n*(n-1))/V + b*k3*V)
    mu_fn = lambda n: ((k2*n*(n-1)*(n-2))/V**2 + n*k4)

    # stochastic matrix Q of dimension NxN
    Q = TridiagA(lambda_fn, mu_fn, N_exact)
    
    # semi-posotive definite transformation
    Qspd = np.dot(Q,Q.T)
    # print("Matrix (V=",np.round(V,2),"): \n\n",Qspd,"\n\n")

    # computing the nth and mth eigenvalues and eigenvectors
    eig0, vec0,eig1,vec1 = find_eigenvalues_eigenvectors(Qspd, n=1, m=2, hermitian=True)
    
    # storing eigenvalues and eigenvectors for different volumes
    exact_eigenvalue0[i] = np.real(eig0)
    exact_eigenvalue1[i] = np.real(eig1)

    i += 1
#*********************************************
print("Exact eigenvalues (at V =,",np.round(V,3),") are ",np.round(exact_eigenvalue0[i-1],4),
      " and", np.round(exact_eigenvalue1[i-1],4))
```

## Classical Simulation

```python
# volume array
Vn_classical = Vn

# eigenvalue arrays
classical_eigenvalue0 = np.zeros(len(Vn_classical))
classical_eigenvalue1 = np.zeros(len(Vn_classical))

#******************************************
for i, V in enumerate(Vn_classical):
    # birth and death rates
    lambda_fn = lambda n: ((a*k1*n*(n-1))/V + b*k3*V)
    mu_fn = lambda n: ((k2*n*(n-1)*(n-2))/V**2 + n*k4)

    # stochastic matrix Q of dimension NxN
    Q = TridiagA(lambda_fn, mu_fn, N)
    
    # semi-posotive definite transformation
    Qspd = np.dot(Q,Q.T)
    # print("Matrix (V=",np.round(V,2),"): \n\n",Qspd,"\n\n")

    # computing the nth and mth eigenvalues and eigenvectors
    eig0, vec0,eig1,vec1 = find_eigenvalues_eigenvectors(Qspd, n=1, m=2, hermitian=True)
    
    # storing eigenvalues and eigenvectors for different volumes
    classical_eigenvalue0[i] = np.real(eig0)
    classical_eigenvalue1[i] = np.real(eig1)

    i += 1
#*********************************************
print("Classical eigenvalues (at V =,",np.round(V,3),") are ",np.round(classical_eigenvalue0[i-1],4),
      " and", np.round(classical_eigenvalue1[i-1],4))
```

## Quantum Simulation

### Pauli Decomposition
Here, we decompose the classical Hermitian Scholgl matrix into Pauli operators.

```python
# transform hermitian matrix to Pauli operator
qubitOp = SparsePauliOp.from_operator(Qspd)

# we can verify the number of qubits in the operator
# It should be the same as 'operator_num_qubits'
print(f"Number of qubits in operator: {qubitOp.num_qubits}")
```

### Ansatz: Trial Wavefunction
Next, we define our trial wavefunction as a quantum circuit using the `TwoLocal` function from the `qiskit.circuit` library. The specific anasatz we use here is similar to the one used for the Hydrogen molecule. So, we are not too optimistic about our results. Let's see what such an ansatz gives us...

```python
# number of qubits for the ansatz: how sophisticated do you want it to be?
ansatz_num_qubits = operator_num_qubits

# trial wavefunction
# ansatz = efficient_su2(ansatz_num_qubits,
#                   rotation_blocks=["ry"], 
#                   entanglement_blocks="cz",
#                   reps=1
#                  )

# ansatz = RealAmplitudes(ansatz_num_qubits, entanglement='linear', reps=1)

ansatz = TwoLocal(ansatz_num_qubits,
rotation_blocks=["ry"],
entanglement_blocks="cx",
entanglement="circular",
reps=4
)


# draw circuit
ansatz.decompose().draw(output='mpl',filename='ansatz.pdf')
```

```python
# See qiskit documentation for more info on optimizers
optimizer = L_BFGS_B(maxiter=500)

# initialize estimator, sampler, and fidelity
estimator = Estimator()

# sampler
sampler = Sampler()

# fidelity
fidelity = ComputeUncompute(sampler)

# number of states to compute and beta parameters
k = 2

# define callbacks to store intermediate steps
counts = []
values = []
steps = []

def callback(eval_count, params, value, meta, step):
    counts.append(eval_count)
    values.append(value)
    steps.append(step)

# instantiate vqd
vqd = VQD(estimator, fidelity, ansatz, optimizer, k=k, callback=callback)
result = vqd.compute_eigenvalues(operator = qubitOp)
vqd_values = result.eigenvalues
```

```python
# print the eigenvalues
print("Quantum eigenvalues (at V = ",np.round(V,3),") are ",np.round(vqd_values[0],4),
      " and", -np.round(np.sqrt(vqd_values[1]),4))
```

```python
print(vqd_values)
```

### VQD Intermediate Steps
Here is a sample of the VQD intermediate steps.

```python
# plotting intermediate steps
plt.rcParams["figure.figsize"] = (12, 8)

steps = np.asarray(steps)
counts = np.asarray(counts)
values = np.asarray(values)

#***********************************************

for i in range(1,3):
    _counts = counts[np.where(steps == i)]
    _values = values[np.where(steps == i)]
    plt.plot(_counts, _values, label=f"State {i-1}",linewidth=2.5)

plt.xlabel("Eval count", fontsize=25)
plt.ylabel("$\lambda$", fontsize=25)
plt.legend(loc="best", fontsize=25)
plt.show()
#***********************************************
```

```python
# volume array
Vn_quantum = Vn

# eigenvalue arrays
quantum_eigenvalue0 = np.zeros(len(Vn_quantum))
quantum_eigenvalue1 = np.zeros(len(Vn_quantum))

#******************************************
for i, V in enumerate(Vn_quantum):
    # birth and death rates
    lambda_fn = lambda n: ((a*k1*n*(n-1))/V + b*k3*V)
    mu_fn = lambda n: ((k2*n*(n-1)*(n-2))/V**2 + n*k4)

    # stochastic matrix Q of dimension NxN
    Q = TridiagA(lambda_fn, mu_fn, N)
    
    # semi-posotive definite transformation
    Qspd = np.dot(Q,Q.T)
    
    # ********************************************************************
    # Quantum computing of eigenvalues
    # ********************************************************************
    
    # transform hermitian matrix to Pauli operator
    qubitOp = SparsePauliOp.from_operator(Qspd)
    
    # instantiate vqd
    vqd = VQD(estimator, fidelity, ansatz, optimizer, k=k)
    result = vqd.compute_eigenvalues(operator = qubitOp)
    vqd_values = result.eigenvalues
    #**********************************************************************
    
    # storing eigenvalues and eigenvectors for different volumes
    quantum_eigenvalue0[i] = np.round(vqd_values[0])
    quantum_eigenvalue1[i] = vqd_values[1]

    i += 1
```
