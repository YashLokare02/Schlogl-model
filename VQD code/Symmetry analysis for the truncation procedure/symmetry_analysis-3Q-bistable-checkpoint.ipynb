#### Creating and Classically Diagonalizing the Schlogl Operator

```python
#******************************************
# volume
V = 8.5

# birth and death rates
lambda_fn = lambda n: ((a*k1*n*(n-1))/V + b*k3*V)
mu_fn = lambda n: ((k2*n*(n-1)*(n-2))/V**2 + n*k4)

# stochastic matrix Q of dimension NxN
Q = TridiagA(lambda_fn, mu_fn, N)

# semi-posotive definite transformation
Qh1 = np.dot(Q,Q.T)
Qh2 = block_diagonal_matrix(Q)

# computing eigenvalues and eigenvectors
eigenvalues, eigenvectors = sp.linalg.eig(Q)
eigenvalues_h1, eigenvectors_h1 = sp.linalg.eigh(Qh1)
eigenvalues_h2, eigenvectors_h2 = sp.linalg.eigh(Qh2)

#*********************************************
smallest_eig, smallest_indices = get_n_min_magnitude_values(eigenvalues, 2)
smallest_eig_h1, smallest_indices_h1 = get_n_min_magnitude_values(eigenvalues_h1, 2)
smallest_eig_h2, smallest_indices_h2 = get_n_min_magnitude_values(eigenvalues_h2, 4)

# print output
print("\n************************************************************************************\n")
print(f"\nSmallest n = {2} eigenvalues and indices:\n\n{smallest_eig}\n\n{smallest_indices}")
print("\n************************************************************************************\n")
print(f"\nSmallest n = {2} eigenvalues_h1 and indices_h1:\n\n{smallest_eig_h1}\n\n{smallest_indices_h1}")
print("\n************************************************************************************\n")
print(f"\nSmallest n = {4} eigenvalues_h2 and indices_h2:\n\n{smallest_eig_h2}\n\n{smallest_indices_h2}")
print("\n************************************************************************************\n")
```

## Custom VQD Algorithm Using StateVector

```python
from scipy.optimize import minimize
from qiskit import QuantumCircuit
from qiskit.circuit.library import TwoLocal
from qiskit.quantum_info import Statevector

def vqd_cost(variational_params, ansatz, known_first_state, operator, penalty_factor):

    # assigning parameters to circuit
    trial_circuit = ansatz.assign_parameters(variational_params)
    
    # Prepare the statevector trial state
    trial_state = Statevector.from_instruction(trial_circuit)
    
    # Compute the overlap of the trial state with the known first state
    overlap = np.abs(np.vdot(known_first_state, trial_state))**2

    # Calculate the expectation value
    expectation_value = trial_state.expectation_value(operator)

    # cost that penalizes the known first state
    vqd_cost = expectation_value + penalty_factor * overlap
    
    # Return the expectation value plus a penalty term proportional to the overlap
    return vqd_cost

# Define a wrapper function for the optimizer
def vqd_wrapper(variational_params, ansatz, known_first_state, operator, penalty_factor):
    cost = vqd_cost(variational_params, ansatz, known_first_state, operator, penalty_factor)
    costs.append(cost)
    return cost

#*********************************************************************************************

# store costs here
costs = []

# defining operator
operator = qubitOp

# defining ansatz
reference_circuit = QuantumCircuit(operator.num_qubits)
reference_circuit.x(0)

variational_form = TwoLocal(
    operator.num_qubits,
    rotation_blocks=["ry"],
    entanglement_blocks="cx",
    entanglement="linear",
    reps=2,
)
ansatz = reference_circuit.compose(variational_form)

ansatz.decompose().draw('mpl')

# initializing the circuit to store our known state
known_circuit = QuantumCircuit(operator.num_qubits)

# known first state
known_first_state = eigenvectors_h1[:,0]

# known state as statevector
known_circuit.initialize(known_first_state, [i for i in range(operator.num_qubits)])
known_first_state = Statevector.from_instruction(known_circuit)

# initializing circuit parameters
num_params = len(ansatz.parameters)
penalty_factor = 1000  # penalty must be larger than gap between eigenvalues
initial_parameters = np.random.rand(num_params)

# Call the optimizer
result = minimize(vqd_wrapper, initial_parameters, args=(ansatz, known_first_state, operator, penalty_factor), method='BFGS')

# The optimal parameters are stored in result.x
optimal_parameters = result.x
```
