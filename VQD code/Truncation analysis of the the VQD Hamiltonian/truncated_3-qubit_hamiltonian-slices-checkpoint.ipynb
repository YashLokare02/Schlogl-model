## Introduction
The cost of the of the VQD algorithm grows with the number of terms in the transformed Pauli operator. To alliate the cost, we truncate the full operator and analyze the accuracy. By adding and removing particular terms in the qubit operator, we systematically study the relevance and weight of each term to the computed eigenvalues.

### Defining Problem
Recall that the VQD finds the $k^{\text{th}}$ lowest (including negative values) eigenvalues. To return sensible results (i.e., eigenvalues of observable quantities), the Schlogl operator must be transformed into a positive-definite hermitian matrix using the transformation
\begin{equation}
Q_{\mathrm{spd}} = Q\cdot Q^T,
\end{equation}
whose eigenvalues are squares of the original matrix's eigenvalues, such that
$$
\lambda^{Q}_{i} = \sqrt{\lambda^{Q_{\mathrm{spd}}}_{i}}
$$
where $\lambda^{Q}_{i}$ are the eigenvalues of the original Schlogl operator and $\lambda^{Q_{\mathrm{spd}}}_{i}$ are the eigenvalues of the transformed semi-positive definite matrix.

### Model Parameters

```python
# model parameters
a = 1
b = 1
k1 = 3
k2 = 0.6
k3 = 0.25
k4 = 2.95
V = 2.5

# number of qubits
operator_num_qubits = 3
print("Initializing...\n\n","-->",str(operator_num_qubits)+"-qubit matrix")

# Matrix size should be 2^num_qubits
N = 2**operator_num_qubits
print(" --> Matrix size = ",N,"x",N,"\n...")
```

## Classical Simulation

```python
# birth and death rates
lambda_fn = lambda n: ((a*k1*n*(n-1))/V + b*k3*V)
mu_fn = lambda n: ((k2*n*(n-1)*(n-2))/V**2 + n*k4)

# stochastic matrix Q of dimension NxN
Q = TridiagA(lambda_fn, mu_fn, N)

# semi-posotive definite transformation
Qspd = np.dot(Q,Q.T)

# computing the nth and mth eigenvalues and eigenvectors
eig0, vec0,eig1,vec1 = find_eigenvalues_eigenvectors(Qspd, n=1, m=2, hermitian=True)

# classical eigenvalues
c_value0 = np.real(eig0)
c_value1 = -np.sqrt(np.abs(np.real(eig1)))

#*********************************************
print("Classical eigenvalues (at V = ",np.round(V,3),") are ",np.round(c_value0,4),
      " and", np.round(c_value1,4))
```

## Quantum Simulation

### Pauli Decomposition
Here, we decompose the classical Hermitian Scholgl matrix into Pauli operators.

```python
# transform hermitian matrix to Pauli operator
qubitOp = SparsePauliOp.from_operator(Qspd)
print(qubitOp)

# we can verify the number of qubits in the operator
# It should be the same as 'operator_num_qubits'
print(f"\n\nNumber of qubits in operator: {qubitOp.num_qubits}")
print("Number of Terms in Original Operator: ", len(qubitOp))
```

### Truncated Pauli Operator

```python
# transform even further (first degree truncation)
qubitOp = qubitOp[0:int(len(qubitOp)/2)]
print(qubitOp)

# length of operator
operator_length = len(qubitOp)

print("\n\nNumber of Terms in truncated Operator: ", operator_length)
```

```python
# sub-operator minsize and maxsize
sub_operator_minsize=9
sub_operator_maxsize=10

# indices representing each term in qubit operator
qubitOp_indices = generate_indices(operator_length)
print("Qubit operator indices:\n ",qubitOp_indices)

# total combinations possible
total_combinations = count_combinations(operator_length)

# every subset in qubit operator
print(f"\nTotal number of sub-operators (truncations): {total_combinations}")

# we only consider some of the subsets, desirably from the smallest
index_subsets = generate_subsets_sorted(qubitOp_indices,
                                        subarray_minsize=sub_operator_minsize,
                                        subarray_maxsize=sub_operator_maxsize)

# truncated combinations
truncated_combinations = len(index_subsets)

print(f"\n ...\n\nConsidering only sub-operators with {sub_operator_minsize} to {sub_operator_maxsize} terms: {truncated_combinations}")
```

### Ansatz: Trial Wavefunction
Next, we define our trial wavefunction as a quantum circuit using the `TwoLocal` function from the `qiskit.circuit` library. The specific anasatz we use here is similar to the one used for the Hydrogen molecule. So, we are not too optimistic about our results. Let's see what such an ansatz gives us...

```python
# number of qubits for the ansatz: how sophisticated do you want it to be?
ansatz_num_qubits = operator_num_qubits

# Define the parameters
a = Parameter('a')
b = Parameter('b')
c = Parameter('c')

# ansatzes
ansatz1 = TwoLocal(ansatz_num_qubits, rotation_blocks=['ry'], entanglement="full",
           entanglement_blocks='cx', reps=1, parameter_prefix='a')
ansatz1.assign_parameters({ansatz1.parameters[i]: a for i in range(len(ansatz1.parameters))})

ansatz2 = RealAmplitudes(ansatz_num_qubits, entanglement='circular', reps=1, parameter_prefix='b')
ansatz2.assign_parameters({ansatz2.parameters[i]: a for i in range(len(ansatz2.parameters))})

ansatz3 = EfficientSU2(num_qubits=ansatz_num_qubits, reps=2, su2_gates=['ry'],
        entanglement='sca', parameter_prefix='c')
ansatz3.assign_parameters({ansatz3.parameters[i]: a for i in range(len(ansatz3.parameters))})

# compose quantum circuit
ansatz = QuantumCircuit(ansatz_num_qubits)

#ansatz.compose(ansatz1, inplace=True)
ansatz.compose(ansatz2, inplace=True)
#ansatz.compose(ansatz3, inplace=True)

# draw circuit
ansatz.decompose().draw(output='mpl')
```

```python
# number of states to compute and beta parameters
k = 2

# define callbacks to store intermediate steps
callback_counts = []
callback_values = []
callback_steps = []

def callback(eval_count, params, value, meta, step):
    counts.append(eval_count)
    values.append(value)
    steps.append(step)

# eigenvalue arrays
quantum_eigenvalue0 = np.zeros(truncated_combinations)
quantum_eigenvalue1 = np.zeros(truncated_combinations)

# keep track of the truncated operators in for loop
truncation_count = 0

#*******************************
for subset in index_subsets:
    
    counts = []
    values = []
    steps = []

    truncated_qubitOp = qubitOp[subset]
    print(f"\nTruncation {truncation_count}:\n ",truncated_qubitOp)
    
    # See qiskit documentation for more info on optimizers
    optimizer = P_BFGS()

    # initialize estimator, sampler, and fidelity
    estimator = Estimator()

    # sampler
    sampler = Sampler()

    # fidelity
    fidelity = ComputeUncompute(sampler)

    # instantiate vqd
    vqd = VQD(estimator, fidelity, ansatz, optimizer, k=k, callback=callback)
    result = vqd.compute_eigenvalues(operator = truncated_qubitOp)
    vqd_values = result.eigenvalues
    
    # store callbacks
    callback_counts.append(counts)
    callback_values.append(values)
    callback_steps.append(steps)
    
    value0 = np.real(vqd_values[0])  # vqd state 0
    value1 = np.real(vqd_values[1])  # vqd state 1

    # Check for invalid values in value0 and assign np.nan
    value0 = np.where(np.logical_or(np.isnan(value0), value0 < 0.0), np.nan, value0)

    # Check for invalid values in value1 and assign np.nan
    value1 = np.where(np.logical_or(np.isnan(value1), value1 < 0.0), np.nan, value1)

    # Calculate the square root for valid values
    quantum_eigenvalue0[truncation_count] = np.where(value0 >= 0.0, -np.sqrt(value0), np.nan)
    quantum_eigenvalue1[truncation_count] = np.where(value1 >= 0.0, -np.sqrt(value1), np.nan)

    
    # next operator
    truncation_count+=1
```

### Eigenvalues from Truncated Operators
Next, let's see how the different truncations affect the accuracy in eigenvalues. We are going to extract the closest N values to classical results.

```python
truncation_result = 0
for idx in range(truncated_combinations):
    print(f"\nVQD Result (Truncation {truncation_result}):\nState 0 = {quantum_eigenvalue0[idx]}\nState 1 = {quantum_eigenvalue1[idx]}")
    truncation_result+=1
```

### Ground State: $\lambda_0$

```python
import matplotlib.pyplot as plt

# plotting intermediate steps
plt.rcParams["figure.figsize"] = (12, 8)

#***********************************************

linestyles = ["-", "--", ":", "-."]  # Different line styles

idx = 0
for closest_indices in closest_indices0:
    linestyle = linestyles[idx % len(linestyles)]  # Cycle through line styles
    
    counts = callback_counts[closest_indices]
    counts = np.asarray(counts)
    
    steps = callback_steps[closest_indices]
    steps = np.asarray(steps)
    
    # in steps, replace unwanted state with zero
    ones = replace_digit_with_zero(steps, 2)
    
    values = callback_values[closest_indices]
    values = np.asarray(values)
    
    # split counts and values to extract desired state
    state0_counts, _, state0_values, _ = split_counts_values(counts, values)
    
    # plotting state 0
    state0_counts = np.array(state0_counts)
    state0_values = np.array(state0_values)

    mask = np.isin(steps, ones)  # Mask for matching elements in steps and twos
    _indices = np.where(mask)[0]  # Indices of matching elements

    # Use the indices to extract the corresponding counts and values
    counts_length = min(len(state0_counts), len(_indices))
    values_length = min(len(state0_values), len(_indices))

    _counts = state0_counts[:counts_length]
    _values = state0_values[:values_length]

    plt.plot(_counts, -np.sqrt(np.abs(_values)), 
             color=f"C{idx}", ls=linestyle, label=f"Truncation {closest_indices} (Terms = {len(qubitOp[index_subsets[closest_indices]])})")
        
    idx+=1

# target state    
plt.axhline(y=c_value0, color="tab:red", ls="--", lw=0.5, label="Target $\lambda_0$")

plt.xlabel("Eval count", fontsize=20)
plt.ylabel("$\lambda_0$", fontsize=20)
plt.legend(loc="best", fontsize=10)
plt.show()
#***********************************************
```
