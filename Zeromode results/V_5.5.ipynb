```python
# Importing relevant libraries
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import matplotlib.pyplot as plt
```

# Comparison plots and RMS difference for all cases (for V = 5.5)

```python
## Import the classical zeromode
zeromode_classic = [[-0.02978204], [-0.02978204], [-0.02978204], [-0.02978204], [ 0.90161196], [ 0.42024286], [ 0.09793796], [ 0.02692673]]
```

```python
# Generate the basis size array
x_axis = []

for i in range(8):
    x_axis.append(i)
```

```python
# Function to compute the RMS difference
def rms_diff(y_op, y):
    
    # Flatten lists
    y_op = [item for sublist in y_op for item in sublist]
    y = [item for sublist in y for item in sublist]
    
    # Extract relevant data for calculation
    y_op = y_op[4:]
    y = y[4:]
    
    y_diff = [(y_op[i] - y[i]) ** 2 for i in range(len(y))]
    
    return np.sqrt(np.mean(y_diff))

# Function to generate the plots for the comparison and error difference
# def generate_plots(x_axis, zeromode_classic, zeromode_qpe):
    
#     # Flatten lists
#     zeromode_classic = [item for sublist in zeromode_classic for item in sublist]
#     zeromode_qpe = [item for sublist in zeromode_qpe for item in sublist]
    
#     # Visualization
#     fig, ax = plt.subplots()
#     ax.plot(x_axis, zeromode_classic, "o-.", 
#             label = 'Classical diagonalization')
#     ax.plot(x_axis, zeromode_qpe, "^--", 
#             label = 'QPE + VQSVD')
#     plt.xlabel('Basis size', fontsize = 14)
#     plt.ylabel('PDF', fontsize = 14)
#     plt.grid(True)
#     leg = plt.legend(frameon = True)
#     leg.get_frame().set_edgecolor('k')
    
#     # Error plot
#     err_plot = [np.abs((zeromode_classic[i] - zeromode_qpe[i])) for i in range(len(zeromode_classic))]

#     # Visualization
#     plt.figure(dpi = 150)
#     plt.xlabel("Basis size", fontsize = 14)
#     plt.ylabel("$|P_{\mathrm{classical}} - P_{\mathrm{quantum}}$|", fontsize = 14)
    
#     # Add grid points
#     plt.grid(True)

#     plt.plot(x_axis, err_plot, '^--')
#     plt.show()

def generate_plots(x_axis, zeromode_classic, zeromode_qpe):
    
    # Flatten lists
    zeromode_classic = [item for sublist in zeromode_classic for item in sublist]
    zeromode_qpe = [item for sublist in zeromode_qpe for item in sublist]
    
    # Convert all (extraneous) negative values to positive values (as these are PDFs)
    zeromode_classic = [np.abs(value) for value in zeromode_classic]
    zeromode_qpe = [np.abs(value) for value in zeromode_qpe]
    
    # Main plot
    fig, ax = plt.subplots()
    ax.plot(x_axis, zeromode_classic, "o-.", label='Classical diagonalization')
    ax.plot(x_axis, zeromode_qpe, "^--", label='QPE + VQSVD')
    ax.set_xlabel('Basis size', fontsize = 14)
    ax.set_ylabel('PDF', fontsize = 14)
    ax.set_yscale('log')
    ax.grid(True)
    leg = ax.legend(frameon = True)
    leg.get_frame().set_edgecolor('k')
    
    # Error plot
    log_classic = [np.log(value) for value in zeromode_classic]
    log_quantum = [np.log(value) for value in zeromode_qpe]
    err_plot = [np.abs(log_classic[i] - log_quantum[i]) for i in range(len(log_classic))]

    # Inset plot
    inset_ax = inset_axes(ax, width = "30%", height = "20%", loc = 'lower left', \
                          bbox_to_anchor = (0.1, 0.2, 1, 1), bbox_transform = ax.transAxes)
    inset_ax.plot(x_axis, err_plot, '^--')
    inset_ax.set_xlabel("Basis size", fontsize = 8)
    inset_ax.set_ylabel("$|P_{\mathrm{classical}} - P_{\mathrm{quantum}}|$", fontsize = 8)
    inset_ax.grid(True)  # Enable grid lines on the inset plot
    
    plt.show()
```

## In the absence and presence of DD

### In the absence of DD

```python
zeromode_qpe = [[-0.15064175], [-0.15050615], [-0.15061936], [-0.15065551], [ 0.8089274 ], [ 0.37695244], [ 0.0879932 ], [ 0.02422124]]

# Generate comparison and error plots
generate_plots(x_axis, zeromode_classic, zeromode_qpe)
```

```python
# Compute the RMS difference
rms_difference = rms_diff(zeromode_classic, zeromode_qpe)

# # Append to relevant array
# RMS_difference.append(rms_difference)

# Print the RMS difference
print('The RMS difference between the classical and quantum zeromodes is:')
print(rms_difference)
print()
```
