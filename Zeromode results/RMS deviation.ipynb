```python
import numpy as np
import matplotlib.pyplot as plt
```

## Visualization

```python
def generate_plot(RMS_difference):
    # Function to generate scatter plots to compare RMS difference values
    
    tick_labels = ['No DD', 'Hahn-X', 'Hahn-Y', 'CP', 'CPMG', 'XYXY', 'YZYZ', 'XZXZ', 'CDD', 'XY8', 'XY16', 'Uhrig-X', \
                   'Uhrig-Y', 'KDD']

    # Setting plot dimensions
    f = plt.figure(dpi = 150)
    f.set_figwidth(15)
    f.set_figheight(4)

    # Set custom tick labels
    plt.xticks(range(len(tick_labels)), tick_labels)

    # Plot 
    plt.scatter(range(len(tick_labels)), RMS_difference, marker = 'D', color = 'blue')

    plt.xlabel('DD sequences', fontsize = 17)
    plt.ylabel("RMS difference [%]", fontsize = 17)

    plt.legend(loc = 'center')
    plt.grid(True) # add grid lines

    plt.show()
```
